```python
import pandas as pd

with pd.HDFStore(r"C:\Users\furza\Downloads\METR-LA.h5\METR-LA.h5", mode = "r") as store:
    df = store["/df"]
    print(store.keys())

    print(df.shape)
    print(df)
```

```python
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
import pickle
import os
os.makedirs("sensor_models",exist_ok=True)

for i in range(0,207):
    model = LinearRegression()
    under_study = df.iloc[:,i].astype(float).dropna().values

    inp = under_study[:-1].reshape(-1,1)
    tgt = under_study[1:].reshape(-1,1)  

    inp_train, inp_test, tgt_train, tgt_test = train_test_split(inp, tgt, test_size=0.2,shuffle=False)

    model.fit(inp_train,tgt_train)
    tgt_predicted = model.predict(inp_test)

    MAE = mean_absolute_error(tgt_test,tgt_predicted)

    with open(f"sensor_models/sensor_{i}.pkl","wb") as f:
        pickle.dump(model,f)
print("All 207 models saved with pickle.")
```
