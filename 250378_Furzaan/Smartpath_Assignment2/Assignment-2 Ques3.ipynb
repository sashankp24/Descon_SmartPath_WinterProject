```python
import pandas as pd

with pd.HDFStore(r"C:\Users\furza\Downloads\METR-LA.h5\METR-LA.h5", mode = "r") as store:
    df = store["/df"]

    print(df.shape)
    print(df.head())
```

```python
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
requirement = input("enter sensor code")
under_study = df[f"{requirement}"].astype(float).values

inp = under_study[:-1].reshape(-1,1) # inputs are the entries starting from 1st to 2nd last in column in under_study
tgt = under_study[1:].reshape(-1,1)  # target are the speeds starting from 2nd to last in column in under_study

inp_train, inp_test, tgt_train, tgt_test = train_test_split(inp, tgt, test_size=0.2,shuffle=False)

model = LinearRegression()
model.fit(inp_train,tgt_train)

tgt_predicted = model.predict(inp_test)

MAE = mean_absolute_error(tgt_test,tgt_predicted)
print(f"MAE for sensor={requirement} is ", MAE)
```
