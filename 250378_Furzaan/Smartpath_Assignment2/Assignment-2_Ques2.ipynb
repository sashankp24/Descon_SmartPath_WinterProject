```python
import numpy as np
import pandas as pd

np.random.seed(42)
# Number of students
n_students = 20

# Columns
hrs = np.random.uniform(0,10,size=n_students) # hrs is hours studied
attendance = np.random.uniform(50,100,size=n_students) # attendance percentage
final_score = 3*hrs + 0.2*attendance + np.random.normal(0,5,size=n_students) # final_score = linear expression + random error
final_score = np.clip(final_score,0,100) # Used to keep score in [0,100]

# Defining Dataframe
df = pd.DataFrame({"Hours_studied":hrs,"Attendance_percentage":attendance,"Final_Score":final_score})
df.head()
```

```python
x = df[["Hours_studied","Attendance_percentage"]]
y = df["Final_Score"]

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Train-test-split (80% train, 20% test)
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state=1)

# Creating a model
reg = LinearRegression()
reg.fit(x_train,y_train)

# Coefficients
print(f"[coeff. for Hours_studied, coeff. for Attendance_Percentage] = {reg.coef_}, intercept = {reg.intercept_}")

# Predictions on test data set
y_predicted = reg.predict(x_test)

# Mean Squared Error
mse = mean_squared_error(y_test,y_predicted)
print("MSE:",mse)

# Model Accuracy
Accuracy = 100*reg.score(x_test,y_test)
print("Accuracy =",Accuracy)
    
```
