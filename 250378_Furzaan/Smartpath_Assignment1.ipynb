Task 1:Python Setup & Data Creation

```python
city_name= input("enter city name")
no_of_road_segments=int(input("enter number of segments"))
avg_speed=float(input("enter avg. speed for roads."))

print(f"In {city_name},there are {no_of_road_segments} road segments with an average speed of {avg_speed} km/h.")
```

```python
speeds=[25,40,32,18.5,50.2,29.3,35.5]
```

Task 2: Python Logic on Traffic Data


```python
for value in speeds:
    print(value)
```

```python
for value in speeds:
    if value < 30:
        print(value)
```

```python
i=0
for value in speeds:
    if value < 30:
        i+=1
print(i,"roads are traffic prone.")
```

Task 3: NumPy-Speed & Time Computation

```python
import numpy as np
arr=np.array(speeds)
```

```python
mean=np.mean(arr)
print(f"The mean speed is {mean} km/h.")
min_speed=np.min(arr)
print(f"The minimum speed is {min_speed} km/h.")
max_speed=np.max(arr)
print(f"The maximum speed is {max_speed} km/h.")
```

```python
travel_time=(2/arr)*60
print(f"Time taken to traverse a road with driving speeds {arr} km/h are {travel_time} minutes respectively.")
```

Task 4: Pandas-Traffic DataFrame Operations 

```python
import pandas as pd
road_ids=range(1,len(speeds)+1)
df=pd.DataFrame({
    "Road_ID": road_ids,
    "Speed (in km/h)":speeds,
    "Lenght (in km)": [2]*len(speeds)

})
print(df)
```

```python
df["Travel Time (in mins)"]= travel_time
df1=df.sort_values(by="Travel Time (in mins)")
print(df1)
```

```python
df1["Traffic level"] = "Low"
df1.loc[df1["Speed (in km/h)"] < 30, "Traffic level"] = "High"
print(df1)
```

Task 5: Matplotlib- Traffic Visualization

```python
import matplotlib.pyplot as plt
plt.plot(road_ids,speeds, marker="o",mec="red",mfc="red")
plt.xlabel("Road IDs")
plt.ylabel("Speeds(in km/h)")
plt.title("Speeds V/S Road IDs ")
plt.grid()
plt.show()
```

```python
plt.bar(road_ids,travel_time, width=0.5)
plt.xlabel("Road IDs")
plt.ylabel("Travel Time (in mins)")
plt.grid()
plt.show()
```

```python
plt.scatter(travel_time,speeds,color="m")
plt.xlabel("Travel time (in mins)",color="red")
plt.ylabel("Speeds (in km/h)",color="red")
plt.title("Speeds v/s Travel Time",color="red")
plt.xticks([2,2.5,3,3.5,4,4.5,5,5.5,6,6.5])
plt.show()
```
