```python
import pandas as pd
url = "https://docs.google.com/spreadsheets/d/1a3MeBgnrY0HkWWCPeLIfjulay6xAOw4T_ehsvgiWz10/export?format=csv&gid=1836256045"
df=pd.read_csv(url)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df[["Hours Studied","Attendence Percentage"]],df["Final Score"],train_size=0.8,random_state=42)
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
model.fit(x_train,y_train)
```

```python

print("cofficients=",model.coef_)
print("intercept=",model.intercept_)
```

```python
from sklearn.metrics import mean_squared_error
y_predicted=model.predict(x_test)
print("MSE=",mean_squared_error(y_test,y_predicted))
```

```python
import pandas as pd
new_url = "https://docs.google.com/spreadsheets/d/1x_5cP1gj8fz94z2vLT3va19JVjBFkSZLalytVPln72s/export?format=csv&gid=1492882329"
df=pd.read_csv(new_url)
stats=df.describe()
print(stats)
```

```python
import matplotlib.pyplot as plt
df.hist(figsize=(14,8)   ,linewidth=2,
    edgecolor='black',
    color='skyblue')
plt.show()
```

```python
x=df.drop(df[["Pass_Class"]],axis=1)
y=df.Pass_Class
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df[["Quiz_Average (0–100)","Time_Spent_Minutes (Daily Avg)","Problems_Solved (Total)"]],df["Pass_Class"],test_size=0.2,random_state=42)
```

```python
from sklearn import linear_model
model=linear_model.LogisticRegression()
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix,ConfusionMatrixDisplay

y_predicted=model.predict(x_test)

print("acuraccy=",accuracy_score(y_test,y_predicted))
print("precision=",precision_score(y_test,y_predicted))
print("recall=",recall_score(y_test,y_predicted))
print("f1 score=",f1_score(y_test,y_predicted))
print("confusion matrix: ",confusion_matrix(y_test,y_predicted))
ConfusionMatrixDisplay.from_predictions(y_test,y_predicted)
```
