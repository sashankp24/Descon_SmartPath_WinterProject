```python
import pandas as pd
table = {
    "Road_ID": ['city_name0001', 'city_name0002', 'city_name0003','city_name0004'],
    "Speed(m/s)": (speed),
    "Length": ['len001','len002','len003','len004']
}
table = pd.DataFrame(table)
table ['Travel_time(s)'] = (travel_time)
table = table.sort_values(by='Travel_time(s)')
table ['Traffic_level'] = ['Low','Mild','Mild','High']
print(table)
```

Task 5

```python
import matplotlib.pyplot as plt
plt.plot(table['Speed(m/s)'],table['Road_ID'])
plt.xlabel('Speed(m/s)')
plt.ylabel('Road_ID')
plt.title('Speed vs Road_ID')
plt.show()
plt.bar(table['Travel_time(s)'],table['Road_ID'])
plt.xlabel('Travel_time(s)')
plt.ylabel('Road_ID')
plt.title('Travel_time vs Road_ID')
plt.show()
```
