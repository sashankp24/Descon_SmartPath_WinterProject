```python
import pandas as pd
data = pd.read_csv('dataset.csv')
data
```

```python
x = data.drop('Final_Score',axis='columns')
x
y = data['Final_Score']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
coeff = model.coef_
intercept = model.intercept_
print(f"coefficient:- {coeff}")
print(f"intercept:- {intercept}")
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_predict,y_test)
mse
```

```python
acc= model.score(x_test,y_test)
print(f"Accuracy:- {acc}")
```

```python
import pandas as pd
```

```python
data = pd.read_csv('METR-LA.csv')
data.head()
```

```python
data.describe()
data
```

```python
from numpy import mean
data= data.dropna()
```

```python
data['timestamp'] = data['Unnamed: 0']
data = data.drop('Unnamed: 0',axis='columns')
data
```

```python
sensor_id = data.columns[1]
df = data[['timestamp', sensor_id]]
df.columns = ['timestamp', 'speed']
df
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['hour']  = df['timestamp'].dt.hour
df['days of week'] = df['timestamp'].dt.dayofweek
df
```

```python
X = df[['hour', 'days of week']]
y = df['speed']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
```

```python
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
```

```python
rfr.fit(x_train,y_train)
```

```python
rfr.score(x_test,y_test)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,root_mean_squared_error
```

```python
mae = mean_absolute_error(y_test,y_pred)
rmse = root_mean_squared_error(y_test,y_pred)
```

```python
print('The mean absoulte error is:-', mae)
print('The root mean squared error is:-', rmse)
```
