Task 1

```python
city_name = "Indore"
num_road_segments = 6
avg_speed = 50
print(f"The city name is {city_name}, the number of road segments is {num_road_segments} and the average speed is {avg_speed}.")
speed_on_segments = [30, 15, 45, 40, 30, 20]
```

Task 2

```python
for speed in speed_on_segments:
    print(speed, "Km/hr\n")
```

```python
for speed in speed_on_segments:
    if(speed < 30):
        print(speed, "Km/hr\n")
```

```python
num_traffic_prone = 0
for speed in speed_on_segments:
    if(speed < 30):
        num_traffic_prone += 1
print(f"number of traffic prone roads is: {num_traffic_prone}.")
```

Task 3

```python
import numpy as np
speed_array = np.array(speed_on_segments)
min_speed = np.min(speed_array)
max_speed = np.max(speed_array)
mean_speed = np.mean(speed_array)
print(f"mean speed is: {mean_speed}, maximum speed is {max_speed} and minimum speed is{min_speed}.")
```

```python
#each road is 2km long
time_each_road = 2/speed_array
print("Travel time on each road is(hours): ", time_each_road)
```

Task 4

```python
import pandas as pd
road_ids = [f"R{i+1}" for i in range(len(speed_array))]
road_length_km = 2
df = pd.DataFrame({
    "Road_ID": road_ids,
    "Speed": speed_array,
    "Length": road_length_km   # same length (2 km) for all roads
})
print("Initial DataFrame:")
print(df)
```

```python
df["Travel_Time"] = time_each_road
print("\nWith Travel_Time column:")
print(df)
```

```python
df = df.sort_values(by="Travel_Time", ascending=True)
print("\nSorted by Travel_Time:")
print(df)
```

```python

def classify_traffic(speed):
    if speed < 20:
        return "High"
    elif speed <= 40:
        return "Medium"
    else:
        return "Low"

df["Traffic_Level"] = df["Speed"].apply(classify_traffic)
print("\nWith Traffic_Level column:")
print(df)
```

Task 5

```python
import matplotlib.pyplot as plt
plt.figure()
plt.plot(df["Road_ID"], df["Speed"], marker="o")
plt.title("Speed vs Road ID")
plt.xlabel("Road ID")
plt.ylabel("Speed (km/h)")
plt.grid(True)
plt.show()
```

```python
plt.figure()
plt.bar(df["Road_ID"], df["Travel_Time"])
plt.title("Travel Time vs Road ID")
plt.xlabel("Road ID")
plt.ylabel("Travel Time (hours)")
plt.grid(True, axis="y")
plt.show()
```
