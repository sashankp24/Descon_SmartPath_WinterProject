Task 1: Python Setup & Data Creation

```python
city_name=input()
n_roads=int(input())
avg_sp=float(input())
print(f"The city of {city_name} has {n_roads} road segments, with an average speed of {avg_sp} km/h")
speeds=[20,32,41,26,30,29]
```

Task 2: Python Logic on Traffic Data

```python
n=len(speeds)
for i in range(n):
  print(speeds[i])
traffic_prone=0
for i in range(n):
  if speeds[i]<30:
    print(speeds[i])
    traffic_prone+=1
print(traffic_prone)
```

Task 3: NumPy – Speed & Time Computation

```python
import numpy as np
sp_arr=np.array(speeds)
mean=np.mean(sp_arr)
min=np.min(sp_arr)
max=np.max(sp_arr)
travel_time=2/sp_arr
print("Travel time for each segment in hours:")
print(travel_time)
```

Task 4: Pandas – Traffic DataFrame Operations


```python
import pandas as pd
df=pd.DataFrame({"Road_ID": range(1,len(speeds)+1),
                 "Speed": speeds,
                 "Length":2})
df["Travel_Time"]=travel_time
df_sorted=df.sort_values(by="Travel_Time")
Traffic_level=[]
for i in range(n):
    if speeds[i]<30:
        Traffic_level.append("High")
    elif speeds[i]<40:
        Traffic_level.append("Medium")
    else:
        Traffic_level.append("Low")
df["Traffic_Level"]=Traffic_level
df_sorted=df.sort_values(by="Travel_Time")
df_sorted
```

Task 5: Matplotlib – Traffic Visualization


```python
import matplotlib.pyplot as mpl
Road_ID=df["Road_ID"]
speeds=df["Speed"]
mpl.figure(figsize=(10,6))
mpl.plot(Road_ID,speeds, marker="o")
mpl.title("Speed vs Road_ID")
mpl.xlabel("Road_ID")
mpl.ylabel("Speed")
mpl.grid(True)
mpl.show()
Travel_time=df["Travel_Time"]
mpl.figure(figsize=(11,6))
mpl.bar(Road_ID,Travel_time)
mpl.title("Travel Time vs Road_ID")
mpl.xlabel("Road_ID")
mpl.ylabel("Travel_Time (Hours)")
mpl.grid(True)
mpl.show()
```
