```python
city="Pune"
road_segments=6
avg_speed=45
print(f"{city} has {road_segments} road segments with an average speed of {avg_speed} km/h")
speeds=[40,25,50,28,60,20]
```

```python
for s in speeds:print(s)
for s in speeds:
    if s<30:print(s)
traffic_prone=sum(1 for s in speeds if s<30)
print(traffic_prone)
```

```python
import numpy as np
arr=np.array(speeds)
print(arr.mean(),arr.max(),arr.min())
travel_time=2/arr
print(travel_time)
```

```python
import pandas as pd
df=pd.DataFrame({"Road_ID":range(1,len(arr)+1),"Speed":arr,"Length":[2]*len(arr)})
df["Travel_Time"]=df["Length"]/df["Speed"]
df=df.sort_values("Travel_Time")
df["Traffic_Level"]=df["Speed"].apply(lambda x:"High" if x<30 else("Medium" if x<50 else "Low"))
print(df)
```

```python
import matplotlib.pyplot as plt
plt.plot(df["Road_ID"],df["Speed"])
plt.title("Speed vs Road ID")
plt.xlabel("Road ID")
plt.ylabel("Speed")
plt.grid()
plt.show()
plt.bar(df["Road_ID"],df["Travel_Time"])
plt.title("Travel Time vs Road ID")
plt.xlabel("Road ID")
plt.ylabel("Travel Time")
plt.grid()
plt.show()
plt.scatter(df["Speed"],df["Travel_Time"])
plt.title("Speed vs Travel Time")
plt.xlabel("Speed")
plt.ylabel("Travel Time")
plt.grid()
plt.show()
```
