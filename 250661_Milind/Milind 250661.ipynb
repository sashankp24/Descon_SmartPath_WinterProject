```python
import pandas as pd
df=pd.DataFrame({
    "Road_ID":[1,2,3,4,5,6],
    "Speed":[10,90,40,60,30,70],
    "Length":[2,2,2,2,2,2]
})
df["Travel Time"]=df["Length"]/df["Speed"]
df=df.sort_values(by="Travel Time")
def Travel(Speed):
    if Speed>30:
        return "High"
    else:
        return "Good"
df["Traffic Level"]=df["Speed"].apply(Travel)
print(df)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 3, 4, 5, 6]
y = [10, 90, 40, 60, 30, 70]

plt.subplot(1, 2, 1)  # (rows, columns, position)
plt.plot(x, y)
plt.xlabel("Road ID")
plt.ylabel("Speed")
plt.grid()
plt.title("Speed vs Road ID")

z = [10, 90, 40, 60, 30, 70]
y= [2/i for i in z]  

plt.subplot(1, 2, 2)
plt.bar(x, y)
plt.xlabel("Road ID")
plt.ylabel("Travel Time")
plt.grid()
plt.title("Travel Time vs Road ID")

plt.tight_layout()
plt.show()
```
