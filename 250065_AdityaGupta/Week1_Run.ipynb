Task 1: Python Setup & Data Creation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

city_name = "New Delhi"
num_road_segments = 6
avg_speed = 35 # in km/hr

print(f"The city of {city_name} has {num_road_segments} major road segment with an average speed of {avg_speed} km/hr.")

road_speeds=[28, 20, 50, 38, 42, 32]
```

Task 2: Python Logic on Traffic Data

```python
for a in road_speeds:
    print(a)
```

```python
for a in road_speeds:
    if a<30:
      print(a)
```

```python
Count = 0
for a in road_speeds:
    if a < 30 :
        Count += 1
print ( Count )             
```

Task 3: NumPy – Speed & Time Computation

```python
speed_array = np.array(road_speeds)

mean_speed = np.mean(speed_array)
max_speed = np.max(speed_array)
min_speed = np.min(speed_array)

print(f"Mean speed:{mean_speed}")
print(f"Max speed: {max_speed}")
print(f"Min speed: {min_speed}")

road_length = 2
travel_time = road_length / speed_array
travel_time_mins = travel_time*60
```

Task 4: Pandas – Traffic DataFrame Operations

```python
Data_Frame = pd.DataFrame({
   "Road_ID": range(1, np.size(road_speeds)+1),
    "Speed": road_speeds,
    "Length": [2] * np.size(road_speeds)  
})

Data_Frame["Travel_Time"] = travel_time_mins

Sorted_DataFrame = Data_Frame.sort_values(by="Travel_Time")

def Traffic_level(speed):
    if speed < 30:
        return "High Traffic"
    else:
        return "Low Traffic"
    
Sorted_DataFrame["Traffic Level"] = Sorted_DataFrame["Speed"].apply(Traffic_level)

print(Sorted_DataFrame)
```

Task 5: Matplotlib – Traffic Visualization

Line Plot b/w Speed and Road ID

```python
plt.plot(Data_Frame["Road_ID"], Data_Frame["Speed"], marker='o')
plt.title("Speed vs Road ID")
plt.xlabel("Road ID")
plt.ylabel("Speed (km/h)")
plt.grid(True)
plt.show()
```

Bar plot b/w Travel Time and Road ID

```python
plt.bar(Data_Frame["Road_ID"],Data_Frame["Travel_Time"])
plt.title("Travel_Time vs Road_ID")
plt.xlabel("Road ID")
plt.ylabel("Travel Time (minutes)")
plt.grid(True)
plt.show()
```

Scatter plot b/w Speed and Travel Time

```python
plt.scatter(Data_Frame["Travel_Time"],Data_Frame["Speed"])
plt.title("Speed vs Travel Time")
plt.xlabel("Time Taken (minutes)")
plt.ylabel("Speed (km/h)")
plt.grid(True)
plt.show()
```
