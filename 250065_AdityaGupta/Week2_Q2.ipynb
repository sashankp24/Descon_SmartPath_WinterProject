Part 1:Making the Data Set

```python
import pandas as pd

data = {
    'Hours_Studied': [
        1, 2, 4, 6, 8, 10, 3, 5, 7, 9,
        11, 12, 2.5, 4.5, 6.5, 8.5, 10.5, 3.5, 5.5, 7.5,
        9.5, 1.5, 12.5, 6, 4, 8, 10, 5, 7, 9
    ],

    'Attendance_Percentage': [
        55, 60, 65, 72, 80, 88, 62, 68, 75, 83,
        90, 95, 58, 66, 73, 81, 89, 64, 70, 78,
        85, 57, 96, 74, 67, 82, 91, 69, 76, 84
    ],

    'Final_Score': [
        45, 50, 58, 66, 74, 82, 52, 60, 68, 76,
        85, 90, 48, 56, 64, 72, 80, 54, 62, 70,
        78, 46, 92, 67, 59, 73, 84, 61, 69, 77
    ]
}

df = pd.DataFrame(data)
df.head()
```

Part 2: Splitting the Data Set

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df[['Hours_Studied','Attendance_Percentage']],df['Final_Score'],test_size=0.20)
```

Part 3: Training the Model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train,y_train)

y_predict = model.predict(X_test)
```

```python
import matplotlib.pyplot as plt

#I have made a graph to visualise the actual vs predicted Final Score and made an x=y line to show the correct line

plt.scatter(y_test, y_predict)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)])
plt.xlabel("Actual Final Score")
plt.ylabel("Predicted Final Score")
plt.title("Actual vs Predicted Final Score")
plt.grid()
plt.show()
```

Part 4: Model Coefficients

```python
print("Model Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```

Part 5: Calculating MSE

```python
from sklearn.metrics import mean_squared_error

print("Mean Squared Error:", mean_squared_error(y_test, y_predict))
```

Part 6: Description and Final Say

The linear regression model predicts the final score based on hours studied and attendance percentage of a student.
The model performs reasonably well as the mean squared error is very low as compared to the result.
