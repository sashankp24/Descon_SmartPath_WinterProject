```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Task 1

```python
city_name = "Chicago"
no_of_road_segments = 6
average_speed = 40
```

```python
print("The city name is "+ city_name)
print("The number of road segments are",no_of_road_segments)
print("The average speed is",average_speed)
```

```python
speed_segments = [50,38,42,30,25,45,28,60]
```

Task 2

```python
for x in speed_segments:
  print(x)
```

```python
count = 0
for x in speed_segments:
  if(x<30):
    count += 1
    print(x)
print("Number of segments prone to traffic are",count)
```

Task 3

```python
speed = np.array(speed_segments)
```

```python
mean_speed = np.mean(speed)
max_speed = np.max(speed)
min_speed = np.min(speed)

print(f"Mean speed: {mean_speed}")
print(f"Maximum speed: {max_speed}")
print(f"Minimum speed: {min_speed}")
```

```python
time = (2/speed)*60
print(time,"minutes")
```

Task 4

```python
data = {
    "Speed" : speed,
     "Length" : [2,2,2,2,2,2,2,2]
}
```

```python
df = pd.DataFrame(data)
Road_id = [1,2,3,4,5,6,7,8]
df['Road_id'] = Road_id
df = df.set_index('Road_id')
print(df)
```

```python
df['Travel_Time(min)'] = time
print(df)
```

```python
df = df.sort_values(by='Travel_Time(min)')
print(df)
```

```python
traffic = []
for x in speed:
  if(x < 30):
    traffic.append(0)
  else: traffic.append(1)
df['Traffic_Level'] = traffic
print(df)
```

Task 5

```python
plt.plot(Road_id,speed)
plt.grid()
plt.xlabel("Road_id")
plt.ylabel("Speed")
plt.title("Speed vs Road Id")
```

```python
plt.bar(Road_id,time)
plt.grid()
plt.xlabel("Road Id")
plt.ylabel("Travel Time(min)")
plt.title("Time vs Road Id")
```

```python
plt.scatter(time,speed)
plt.xlabel("Travel Time")
plt.ylabel("Speed")
plt.title("Speed vs Travel Time")
```
