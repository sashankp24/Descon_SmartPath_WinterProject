# Question 2

Importing all the needed libraries

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.metrics import mean_squared_error
```

Creating a dataframe

```python
data = {
    "Hours_Studied": [1, 2, 2.5, 3, 4, 5, 6, 7, 8, 9, 10, 4.5, 6.5, 7.5, 8.5],
    "Attendance_Percentage": [55, 60, 58, 65, 70, 75, 78, 82, 85, 90, 95, 72, 80, 88, 92],
    "Final_Score": [42, 48, 46, 55, 60, 65, 70, 74, 78, 85, 90, 63, 72, 80, 87]
}

df = pd.DataFrame(data)
df
```

Splitting the database and applying the model

```python
x = df[["Hours_Studied", "Attendance_Percentage"]]
y = df["Final_Score"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=21)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)
```

```python
print("Coefficient for Hours Studied:", model.coef_[0])
print("Coefficient for Attendance Percentage:", model.coef_[1])
print("Intercept:", model.intercept_)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R2 Score:", r2)
```

This linear regression model demonstrates how machine learning can be applied to educational platforms like this to predict student performance.

One reason performance could improve is that I have taken a linear database here but real student performance can be differnt and may or may not have a linear relationship with the given factors. And it can also be improved by adding a few more factors to analyse student's prformance.

# Question 3

Loading the dataset and needed libraries.

```python
import h5py
import pandas as pd
import numpy as np

df = pd.read_hdf("METR-LA.h5")
df.head()
```

```python
df.dtypes.head()
```

```python
df.shape
```

```python
df.head(25)
```

```python
traffic_df = pd.DataFrame(index=df.index)

traffic_df["hour"] = df.index.hour
traffic_df["day_of_week"] = df.index.dayofweek
traffic_df["avg_speed"] = df.mean(axis=1)

traffic_df.dropna(inplace=True)
traffic_df.head()
```

```python
x = traffic_df[["hour", "day_of_week"]]
y = traffic_df["avg_speed"]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

Applying linear regression.

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

y_pred = model.predict(x_test)

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("Mean Absolute Error (MAE):", mae)
print("Root Mean Squared Error (RMSE):", rmse)
print("R2 Score:", r2)
```

Applying Gradient Boosting Classifier.

```python
from sklearn.ensemble import GradientBoostingRegressor

gb_model = GradientBoostingRegressor()

gb_model.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

gb_pred = gb_model.predict(x_test)

gb_mae = mean_absolute_error(y_test, gb_pred)
gb_rmse = np.sqrt(mean_squared_error(y_test, gb_pred))
gb_r2 = r2_score(y_test, gb_pred)

print("Gradient Boosting MAE:", gb_mae)
print("Gradient Boosting RMSE:", gb_rmse)
print("Gradient Boosting RÂ²:", gb_r2)
```

Applying random forest model.

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor()

rf_model.fit(x_train, y_train)
```

```python
rf_pred = rf_model.predict(x_test)

rf_mae = mean_absolute_error(y_test, rf_pred)
rf_rmse = np.sqrt(mean_squared_error(y_test, rf_pred))
rf_r2 = r2_score(y_test, rf_pred)

print("Random Forest MAE:", rf_mae)
print("Random Forest RMSE:", rf_rmse)
print("Random Forest RÂ²:", rf_r2)
```

Estimated travel time on a 10 kilometer path by linear regression model output.

```python
route_distance_km = 10 # assuming a distance of 10 kilometers

travel_time_hours = route_distance_km / y_pred
travel_time_minutes = travel_time_hours * 60

print("Estimated Travel Time (minutes):", travel_time_minutes)
```

```python
route_distance_km = 10

travel_time_minutes_gb = (route_distance_km / gb_pred) * 60

print("Estimated Travel Time (Gradient Boosting, minutes):", travel_time_minutes_gb)
```

```python
route_distance_km = 10 # assuming a distance of 10 kilometers

travel_time_minutes_rf = (route_distance_km / rf_pred) * 60

print("Estimated Travel Time (Random Forest, minutes):", travel_time_minutes_rf)
```
