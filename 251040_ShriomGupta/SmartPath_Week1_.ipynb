#Task 1: Python Setup & Data Creation

```python
#creating variables

city_name = 'Lucknow'
num_road_segments = '15'
average_speed = '42'

print(f"{city_name} city has {num_road_segments} road segments.The average speed in {city_name} is {average_speed} km/hr. ")

road_speeds = [40,45,30,60,50,80,15,10,25]
```

#Task 2: Python Logic on Traffic Data

```python
print("All Road Speeds:")
for speed in road_speeds:
  print(speed)

traffic_prone_roads = 0

print("Road speeds less than 30 km/hr:")
for speed in road_speeds:
  if(speed<30):
    traffic_prone_roads += 1
    print(speed)

print(f"{traffic_prone_roads} roads are prone to traffic(speed<30km/hr).")
```

#Task 3: NumPy – Speed & Time Computation

```python
import numpy as np
array_speeds = np.array(road_speeds)

mean_speed = np.mean(array_speeds)
max_speed = np.max(array_speeds)
min_speed = np.min(array_speeds)

print("Mean Speed:", mean_speed)
print("Maximum Speed:", max_speed)
print("Minimum Speed:", min_speed)

road_length = 2
travel_time = road_length / array_speeds * 60

print("Travel time for each road in minutes is:", np.round(travel_time, 2))
```

#Task 4: Pandas – Traffic DataFrame Operations

```python
import pandas as pd

df = pd.DataFrame({"Road_ID": np.arange(1, 10),"Speed": array_speeds,"Length": np.full(9, 2)})
df["Travel_Time"] = df["Length"] / df["Speed"] * 60

df_sorted = df.sort_values(by="Travel_Time",)

df_sorted["Traffic_Level"] = "Low"

df_sorted.loc[df["Speed"] <= 50, "Traffic_Level"] = "Medium"
df_sorted.loc[df["Speed"] < 30, "Traffic_Level"] = "High"

df_sorted
```

#Task 5: Matplotlib – Traffic Visualization

```python
import matplotlib.pyplot as plt

#1 Plot Speed vs Road_ID using a line plot.
plt.figure()
plt.plot(df_sorted["Road_ID"], df_sorted["Speed"])
plt.title("Speed vs Road ID")
plt.xlabel("Road ID")
plt.ylabel("Speed (km/h)")
plt.grid()
plt.show()

#2 Plot Travel_Time vs Road_ID using a bar plot.
plt.figure()
plt.bar(df_sorted["Road_ID"], df_sorted["Travel_Time"])
plt.title("Travel Time vs Road ID")
plt.xlabel("Road ID")
plt.ylabel("Travel Time (hours)")
plt.grid()
plt.show()
```
