```python
#Task 1:
city_name="Mumbai"
number_of_road_segements=12
average_speed=22.5
print(f'In {city_name} with number of segments{number_of_road_segements} and the average speed of vehicles on road is {average_speed}')
Road_segment_speeds=[22.5,55,46,34,26,42]
print(Road_segment_speeds)
```

```python
for speed in Road_segment_speeds:
    print(speed)
print('Road speeds that are less than 30 km/h.')
for speed in Road_segment_speeds:
   if speed<30:
       print(speed)
count=0
for speed in Road_segment_speeds:
   if speed < 30:
       count+=1
print(f"The Total roads that are traffic prone (i.e. speed<30) is: {count}")
```

```python
import numpy as np
Speed_array=np.array(Road_segment_speeds)
Mean_speed=np.mean(Speed_array)
Max_speed=np.max(Speed_array)
Min_speed=np.min(Speed_array)
print(f"The Mean speed is: {Mean_speed}")
print(f"The Max speed is: {Max_speed}")
print(f"The Min speed is: {Min_speed}")
road_length=2
total_time_hrs=road_length/Speed_array
total_time_min=total_time_hrs*60
print(f'The total time in minutes for each segment is:{total_time_min}')

```

```python
import pandas as pd
import numpy as np

Road_ID=[12,24,48,56,72]
Speed=[22.5,50,82,69,12]
Length=[11,22,33,44,55]
speed_array_2=np.array(Speed)
length_array=np.array(Length)
Travel_Time_array=length_array/speed_array_2
Travel_Time_list=[11/22.5,22/50,33/82,44/69,55/12]
Traffic_Level=['20-30','30-60','50-100','50-90','10-40']
Roads=pd.DataFrame({'Road_ID':Road_ID,'Speed':Speed,'Length':Length})
print(Roads)
print("After adding column Travel_Time")
Roads['Travel_Time']=Travel_Time_list
print(Roads)
print(Roads.sort_values(by='Travel_Time', ascending=True,inplace=True))
print("After sorting Travel_Time")
print(Roads)
print("After adding column Traffic_Level")
Roads['Travel_Level']=Traffic_Level
print(Roads)
```

```python
import matplotlib.pyplot as plt
plt.plot(Speed,Road_ID)
plt.xlabel('Speed')
plt.ylabel('Road_ID')
plt.title('Speed vs Road_ID')
plt.grid(True)
plt.show

```

```python
plt.bar(Travel_Time_list,Road_ID,width=0.1)
plt.xlabel('Travel_Time')
plt.ylabel('Road_ID')
plt.title('Travel_Time vs Road_ID')
plt.grid(True)
plt.show()
```
